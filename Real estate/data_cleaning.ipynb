```python
def clean_ticket_price(Real_Estate_df):
    Real_Estate_df['Ticket_Price_Cr'] = (
        Real_Estate_df['Ticket_Price_Cr']
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.extract(r'(\d+\.?\d*)')[0]  # Extract numeric part
        .astype(float)
    )
    return Real_Estate_df
```

```python
def parse_purchase_quarter(Real_Estate_df):
    # If format is "Q1 2022"
    Real_Estate_df['Purchase_Quarter'] = pd.PeriodIndex(Real_Estate_df['Purchase_Quarter'], freq='Q')
    return Real_Estate_df

```

```python
def convert_nri_flag(Real_Estate_df):
    Real_Estate_df['NRI_Buyer'] = Real_Estate_df['NRI_Buyer'].map({'Yes': True, 'No': False})
    return Real_Estate_df
```

```python
def clean_property_data(Real_Estate_df):
    Real_Estate_df = clean_ticket_price(Real_Estate_df)
    Real_Estate_df = parse_purchase_quarter(Real_Estate_df)  # or convert_quarter_to_date(df)
    Real_Estate_df = convert_nri_flag(Real_Estate_df)
    return Real_Estate_df
```
