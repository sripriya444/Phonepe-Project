```python
RealEstate_df['Unit_Size_Sqft'] = pd.to_numeric(RealEstate_df['Unit_Size_Sqft'], errors='coerce')
RealEstate_df['Ticket_Price_Cr'] = pd.to_numeric(RealEstate_df['Ticket_Price_Cr'], errors='coerce')
RealEstate_df['Connectivity_Score'] = pd.to_numeric(RealEstate_df['Connectivity_Score'], errors='coerce')
RealEstate_df['Amenity_Score'] = pd.to_numeric(RealEstate_df['Amenity_Score'], errors='coerce')
RealEstate_df['Locality_Infra_Score'] = pd.to_numeric(RealEstate_df['Locality_Infra_Score'], errors='coerce')
RealEstate_df['Avg_Traffic_Time_Min'] = pd.to_numeric(RealEstate_df['Avg_Traffic_Time_Min'], errors='coerce')
RealEstate_df
```

```python
RealEstate_df.info()
```

```python
cat_cols = ['Micro_Market', 'Project_Name', 'Developer_Name', 'Configuration',
            'Transaction_Type', 'Buyer_Type', 'Possession_Status',
            'Sales_Channel', 'NRI_Buyer', 'Buyer_Comments']
RealEstate_df[cat_cols] = RealEstate_df[cat_cols].astype('category')
```

```python
RealEstate_df['Purchase_Quarter'] = pd.to_datetime(RealEstate_df['Purchase_Quarter'], errors='coerce')
```

```python
RealEstate_df['Micro_Market'] = RealEstate_df['Micro_Market'].str.strip().str.lower().str.title()
```

```python
RealEstate_df['Configuration'] = RealEstate_df['Configuration'].str.upper().str.replace('+', ' PLUS')
```

```python
RealEstate_df['Amenity_Score_Flag'] = RealEstate_df['Amenity_Score'].isna().astype(int)
```

```python
RealEstate_df['Amenity_Score'] = RealEstate_df.groupby('Micro_Market')['Amenity_Score'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
RealEstate_df.drop_duplicates(subset='Property_ID', inplace=True)
```

```python
RealEstate_df = RealEstate_df[(RealEstate_df['Ticket_Price_Cr'] > 0) & (RealEstate_df['Ticket_Price_Cr'] < 100)]
```

```python
RealEstate_df = RealEstate_df[RealEstate_df['Purchase_Quarter'].notna()]
```

```python
RealEstate_df.info()
```

```python
RealEstate_df['Configuration'] = (
    RealEstate_df['Configuration']
    .str.upper()
    .str.replace('+', ' PLUS', regex=False)
    .str.strip()
)
```

```python
RealEstate_df['Micro_Market'] = (
    RealEstate_df['Micro_Market']
    .str.lower()
    .str.strip()
    .str.title()
)
```

```python
if 'No Comment' not in RealEstate_df['Buyer_Comments'].cat.categories:
    RealEstate_df['Buyer_Comments'] = RealEstate_df['Buyer_Comments'].cat.add_categories(['No Comment'])

RealEstate_df['Buyer_Comments'] = RealEstate_df['Buyer_Comments'].fillna('No Comment')
```

```python
RealEstate_df = RealEstate_df[RealEstate_df['Unit_Size_Sqft'].notna()]
```

```python
RealEstate_df = RealEstate_df.rename(columns={'Developer_Name': 'Builder'})
```

```python
RealEstate_df['Price_per_Sqft'] = (
    RealEstate_df['Ticket_Price_Cr'] * 1e7 / RealEstate_df['Unit_Size_Sqft']
)
```

```python
RealEstate_df['Quarter_Number'] = pd.to_datetime(
    RealEstate_df['Purchase_Quarter'], errors='coerce'
).dt.to_period('Q').astype(str)
```

```python
RealEstate_df['Booking_Flag'] = (
    RealEstate_df['Possession_Status'].str.strip().str.lower() == 'ready to move'
).astype(int)
```

```python
def clean_real_estate_data(RealEstate_df):
    RealEstate_df = RealEstate_df.copy()

    # Rename for consistency (if needed)
    RealEstate_df.rename(columns={'Builder': 'Developer_Name'}, inplace=True)

    # Type casting
    RealEstate_df['Amenity_Score_Flag'] = RealEstate_df['Amenity_Score_Flag'].astype(bool)
    RealEstate_df['NRI_Buyer'] = RealEstate_df['NRI_Buyer'].astype(bool)

    # Flagging imputation
    RealEstate_df['Traffic_Imputed'] = RealEstate_df['Avg_Traffic_Time_Min'].isnull()

    # Optional: drop rows with critical nulls
    # RealEstate_df.dropna(subset=['Ticket_Price_Cr', 'Unit_Size_Sqft'], inplace=True)

    return RealEstate_df
```
