```python
import pandas as pd
import json
import os

# Paths
agg_transaction_state_path = "C:/Users/Kamalesh Kumar E C/Desktop/SRIPRIYA/pulse/data/aggregated/transaction/country/india/state"
agg_transaction_state_list = os.listdir(agg_transaction_state_path)

# Initialize dictionary
agg_transaction = {
    'States': [], 'Years': [], 'Quarter': [],
    'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []
}

# Step 1: Parse JSON files
for state in agg_transaction_state_list:
    years = os.listdir(f'{agg_transaction_state_path}/{state}')
    for year in years:
        quarters = os.listdir(f'{agg_transaction_state_path}/{state}/{year}')
        for qtr in quarters:
            path = f'{agg_transaction_state_path}/{state}/{year}/{qtr}'
            try:
                with open(path, 'r') as f:
                    data = json.load(f)
                for txn in data['data']['transactionData']:
                    transaction = txn['name']
                    count = txn['paymentInstruments'][0]['count']
                    amount = txn['paymentInstruments'][0]['amount']

                    agg_transaction['States'].append(state)
                    agg_transaction['Years'].append(int(year))
                    agg_transaction['Quarter'].append(int(os.path.splitext(qtr)[0]))
                    agg_transaction['Transaction_type'].append(transaction)
                    agg_transaction['Transaction_count'].append(count)
                    agg_transaction['Transaction_amount'].append(amount)
            except Exception as e:
                print(f"Error processing {path}: {e}")

# Step 2: Convert to DataFrame
agg_transaction_df = pd.DataFrame(agg_transaction)
agg_transaction_df

```

```python
# Step 3: Sort and compute previous quarter amount
agg_transaction_df.sort_values(by=["States", "Transaction_type", "Years", "Quarter"], inplace=True)
agg_transaction_df["Previous_amount"] = agg_transaction_df.groupby(["States", "Transaction_type"])["Transaction_amount"].shift(1)
agg_transaction_df['Previous_amount'] = agg_transaction_df['Previous_amount'].fillna(0)
agg_transaction_df

```

```python
def detect_trend(current, previous):
    if previous == 0:
        return 'No Data'
    elif current > previous:
        return 'Growth'
    elif current < previous:
        return 'Decline'
    else:
        return 'Stable'

agg_transaction_df['Trend'] = agg_transaction_df.apply(
    lambda row: detect_trend(row['Transaction_amount'], row['Previous_amount']),
    axis=1
)
agg_transaction_df

```
